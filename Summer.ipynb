```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
data = pandas.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("No. of cities which held Summer Olympics are:",len(data['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
sport_gold = data[data['Medal']=='Gold'].groupby('Sport').size().sort_values(ascending=False).head()
plt.figure(figsize=(6,6))
plt.xlabel('Sport')
plt.ylabel('No. of gold medals')
sns.barplot(x=sport_gold.index, y=sport_gold, palette = 'Set1');
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sport_medal = data.groupby('Sport').size().sort_values(ascending=False).head()
plt.figure(figsize=(6,6))
plt.xlabel('Sport')
plt.ylabel('No. of Medals')
sns.barplot(x=sport_medal.index, y=sport_medal, palette = 'Set1');
```

### 4. Which player has won most number of medals? (Top 5)

```python
athlete_db = data.groupby('Athlete').size().sort_values(ascending=False).head()
plt.figure(figsize=(12,6))
plt.xlabel('Athlete')
plt.ylabel('No. of medals')
sns.barplot(x=athlete_db.index, y=athlete_db, palette = 'Set1');
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
most_gold = data[data['Medal']=='Gold'].groupby('Athlete').size().sort_values(ascending=False).head()
plt.figure(figsize=(12,6))
plt.xlabel('Athlete')
plt.ylabel('No. of Gold medals')
sns.barplot(x=most_gold.index, y=most_gold, palette = 'Set1');
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data[['Year']][(data['Country'] == 'IND') & (data['Medal'] == 'Gold')].min()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
players_db = data.groupby('Event').size().sort_values(ascending=False).head()
plt.figure(figsize=(12,6))
plt.xlabel('Event')
plt.ylabel('No. of popular event')
sns.barplot(x=players_db.index, y=players_db, palette = 'Set1');
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
female_gold = data[data['Gender']=='Women'].groupby('Sport').size().sort_values(ascending=False).head()
plt.figure(figsize=(8,6))
plt.xlabel('Event')
plt.ylabel('No. of Female Gold Medalists')
sns.barplot(x=female_gold.index, y=female_gold, palette = 'Set1');
```
